```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_556566/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_556566.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt  # <-- You missed this line

# Bar chart for IncomeGroup
df['IncomeGroup'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Distribution of Countries by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Number of Countries")
plt.show()
```

```python
df['Region'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title("Distribution of Countries by Region")
plt.xlabel("Region")
plt.ylabel("Number of Countries")
plt.xticks(rotation=45)  # rotate labels for readability
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_556566/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_556566.csv")
print(df.head())
```
